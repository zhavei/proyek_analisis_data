```python
import pandas as pd

city_names = ['Jakarta', 'Bandung', 'Makassar', 'Surabaya', 'Medan', 'Yogyakarta', 'Malang']
population = [498044, 964254, 491918, 8398748, 653115, 883305, 744955]
num_airports = [2, 2, 8, 3, 1, 3, 2]

df = pd.DataFrame({
    "City Name": city_names,
    "Population": population,
    "AirPorts": num_airports,
})

df_corr = pd.DataFrame({
    "Population": population,
    "AirPorts": num_airports,
})

df.describe(include="all") #melihat rangkuman parameter statistik (mean, median, dll.) 
df.hist() #dapat digunakan untuk memeriksa distribusi data secara cepat.
df_corr.corr() #memeriksa korelasi antar data numerik pada sebuah DataFrame
```
