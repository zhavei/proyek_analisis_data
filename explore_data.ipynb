## Mengeksplorasi Parameter Statistik dari Data

```python
import pandas as pd

city_names = ['Jakarta', 'Bandung', 'Makassar', 'Surabaya', 'Medan', 'Yogyakarta', 'Malang']
population = [498044, 964254, 491918, 8398748, 653115, 883305, 744955]
num_airports = [2, 2, 8, 3, 1, 3, 2]

df = pd.DataFrame({
    "City Name": city_names,
    "Population": population,
    "AirPorts": num_airports,
})

df_corr = pd.DataFrame({
    "Population": population,
    "AirPorts": num_airports,
})

df.describe(include="all") #melihat rangkuman parameter statistik (mean, median, dll.) 
df.hist() #dapat digunakan untuk memeriksa distribusi data secara cepat.
df_corr.corr() #memeriksa korelasi antar data numerik pada sebuah DataFrame
```

### Mengelompkan data

```python
import pandas as pd
 
body_measurement_df = pd.DataFrame.from_records((
  (2, 83.82, 8.4),
  (4, 99.31, 16.97),
  (3, 96.52, 14.41),
  (6, 114.3, 20.14),
  (4, 101.6, 16.91),
  (2, 86.36, 12.64),
  (3, 92.71, 14.23),
  (2, 85.09, 11.11),
  (2, 85.85, 14.18),
  (5, 106.68, 20.01),
  (4, 99.06, 13.17),
  (5, 109.22, 15.36),
  (4, 100.84, 14.78),
  (6, 115.06, 20.06),
  (2, 84.07, 10.02),
  (7, 121.67, 28.4),
  (3, 94.49, 14.05),
  (6, 116.59, 17.55),
  (7, 121.92, 22.96),
), columns=("age", "height_cm", "weight_kg"))
 
body_measurement_df.groupby(by="age").mean()
body_measurement_df.groupby(by="age").agg({
     'height_cm': 'mean',
    'weight_kg': ['mean','max', 'min'],
})
```

```python
import matplotlib.pyplot as plt
 
cities = ('Bogor', 'Bandung', 'Jakarta', 'Semarang', 'Yogyakarta', 
          'Surakarta','Surabaya', 'Medan', 'Makassar')
 
populations = (45076704, 11626410, 212162757, 19109629, 50819826, 17579085,
               3481, 287750, 785409)
 
plt.bar(x=cities, height=populations)
plt.xticks(rotation=45)
plt.show()

plt.barh(y=cities, width=populations)
plt.show

df = pd.DataFrame({
    'Cities': cities,
    'Population': populations,
})
 
df.sort_values(by='Population', inplace=True)
 
plt.barh(y=df["Cities"], width=df["Population"])
plt.show()
```
